```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic dataset with correlated features
np.random.seed(42)
data = pd.DataFrame({
    'Feature 1': np.random.rand(100),
    'Feature 2': np.random.rand(100) * 0.5 + 0.5,
    'Feature 3': np.random.rand(100) * 2,
    'Feature 4': np.random.rand(100) * 0.8 + 0.2
})

# Compute the correlation matrix
corr_matrix = data.corr()

# Heatmap of the correlation matrix using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# # 4. Using the given x and y data, generate a 3D surface plot to visualize the  function z=sin(x2+y 2)
#  x = np.linspace(-5, 5, 100)
#   y = np.linspace(-5, 5, 100)
#   x, y = np.meshgrid(x, y)
#   z = np.sin(np.sqrt(x**2 + y**2))
#   data = { 'X': x.flatten(),
#           'Y': y.flatten(),
#            'Z': z.flatten() }
#             df = pd.DataFrame(data)
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generating data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Creating a DataFrame for 3D surface plot
data_surface = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df_surface = pd.DataFrame(data_surface)

# Creating a 3D surface plot using Plotly
fig_surface = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Adding title and axis labels
fig_surface.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))', scene = dict(
                    xaxis_title='X',
                    yaxis_title='Y',
                    zaxis_title='Z'),
                    width=800, height=800)

# Display the plot
fig_surface.show()
```

```python
from bokeh.plotting import figure, show
import numpy as np
from bokeh.models import ColumnDataSource

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.rand(100) * 50
colors = np.random.randint(1, 256, 100)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a scatter plot
p = figure(title="Scatter Plot with Random Data", x_axis_label='x', y_axis_label='y')

p.scatter('x', 'y', size='sizes', color='colors', source=source, fill_alpha=0.6)

# Show the plot
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Output to Jupyter notebook
output_notebook()

# Data for bar chart
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Convert data to ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a figure
p = figure(x_range=fruits, plot_height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add bars
p.vbar(x='fruits', top='counts', width=0.9, source=source,
       legend_field="fruits", line_color='white',
       fill_color=factor_cmap('fruits', palette=["#c9d9d3", "#718dbf", "#e84d60", "#ddb7b1"], factors=fruits))

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure for the histogram
p = figure(title="Histogram of Random Data", background_fill_color="#fafafa")

# Plot the histogram as a series of rectangles
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
# data_heatmap = np.random.rand(10, 10)
# x = np.linspace(0, 1, 10)
# y = np.linspace(0, 1, 10)
# xx, yy = np.meshgrid(x, y)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColorBar, LinearColorMapper
import numpy as np

# Generate data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a figure for heatmap
p = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1))

# Set up color mapper
mapper = LinearColorMapper(palette="Viridis256", low=np.min(data_heatmap), high=np.max(data_heatmap))

# Add the heatmap
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=mapper)

# Add color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
